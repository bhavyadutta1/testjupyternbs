```python
import pandas as pd
from matplotlib import pyplot as plt 
```

### Load CSV file

```python
original_data = pd.read_csv('cre1_original_data.csv')
original_data
```

### Remove Columns

```python
original_data.drop(original_data.columns[2:6], axis=1, inplace=True)      # Axis 1 means column and 0 means rows. inplace=true specifies the drop operation to be in same dataframe rather creating a copy of the dataframe after drop.
original_data
```

### Add additional Adjusted time

```python
original_data['adjusted_time'] = original_data['Time'] - original_data.Time.iloc[0]  # Also can be written as original_data['adjusted_time'] = original_data['Time'] - 3.11
original_data
```

### Remove Rows

```python
original_data.drop(original_data.index[0:4], axis=0, inplace=True)
original_data
```

### Reset index list to 0

```python
original_data.reset_index(drop=True, inplace=True)
original_data
```

### Removing Rows

```python
original_data.drop(original_data.index[1590:], axis=0, inplace=True)
original_data
```

### Display full table

```python
#See all rows in the table
#pd.set_option('display.max_rows', None)     ! None means there is no return value !
original_data
```

### Plotting Conductivity vs Time

```python
plt.plot( original_data.adjusted_time ,original_data.lambda2Reactor)
plt.title("Conductiviy vs time")
plt.xlabel("Time [s]")
plt.ylabel("Conductivity {lambda2} [mS/cm]")
plt.legend(["F (ti)"])
plt.show()
```
