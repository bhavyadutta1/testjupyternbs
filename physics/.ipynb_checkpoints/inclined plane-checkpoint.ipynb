## <font color="royalblue"> 1 Theory

The relation between the force that acts on a body and the resulting motion of the body as a consequence
of this force is described by Newton’s second law
<p style="text-align:center"> $$ \vec{F}=m\vec{a} $$  <p style="text-align:right"> (1) </p>

where $\vec{F}$ is the external force acting on the body,$\vec{a}$ the resulting acceleration of the body and m the mass
of the body. According to this equation the acceleration is always in the direction of the external force and
the vectors$\vec{F}$ and $m\vec{a}$ have the same length, which leads to the scalar version of Newton’s second law:
<p style="text-align:center"> $$ F=ma $$  <p style="text-align:right"> (2) </p>
 
with $F = |\vec{F}| \; \text{and  a} = |\vec{a}|$

In the case of free fall the external force is the gravitational force $\vec{F}_G$, which on the scale of a laboratory
can be assumed as a constant, downward pointing force. The magnitude of the gravitational force acting on
the body is proportional to the mass of the body, where the constant of proportionality is $g = 9.81 m/s^2$

$F_G = |\vec{F}_G| = m g$
Newton’s second law can then be written as $\vec{F}_G = m \vec{a}$, or in components:

<p style="text-align:center"> $$\begin{pmatrix}
0 \\
-mg \\
0
\end{pmatrix} = m \begin{pmatrix}
a_x \\
a_y \\
a_z
\end{pmatrix}$$  <p style="text-align:right"> (3) </p>


From this follows, that in free fall there is only a non-vanishing acceleration in y-direction with $a_y = −g$.
Or, reading Newton’s second law in the other direction: if a body with mass m is accelerated downwards in
free fall with an acceleration $\vec{g}$, it experiences a gravitational force $\vec{F}_G = m \vec{g}$.

For a constant external force, i.e. if the force vector does not change in time and in space, the resulting
movement of the body is according to Newton’s second law a linear movement with constant acceleration,
where the acceleration points in the same direction as the force. The corresponding vector version of the
position time relation for a movement with constant acceleration is:

<p style="text-align:center"> $$\vec{x}(t)=\frac{1}{2}\vec{a}t^2 +\vec{v}_0t + \vec{x}_0$$  <p style="text-align:right"> (4) </p>


where $\vec{x_0} = \vec{x}(t = 0)$ is the initial position of the body, $\vec{v_0}$ its initial velocity and $\vec{a} = \vec{F}/m$  its acceleration
due to the external force $\vec{F}$ . In the free fall case with no initial velocity the movement is vertical and is
described by the y-compontent of the position time equation:

<p style="text-align:center"> $$y(t)= -\frac{1}{2}gt^2 + y_2$$  <p style="text-align:right"> (5) </p>


Before stopwatches were available it was very hard to measure the gravitational acceleration g accurately.
For that reason Galileo Galilei thought of a slow motion version of the free fall experiment. This experiment
is the inclined plane, because a body that is moving down an inclined plane is accelerated with an acceleration smaller than g. If the body is accelerated along the inclined plane, then, according to Newton’s second law,
there has to be a force in the same direction, i.e. parallel to the inclined plane. This force is called the slope
force $F_sl$ (see figure 1). But since the only external force in this situation is the gravitational force, the slope
force must be related to the gravitational force.

there has to be a force in the same direction, i.e. parallel to the inclined plane. This force is called the slope
force $\vec{F_sl}$ (see figure 1). But since the only external force in this situation is the gravitational force, the slope
force must be related to the gravitational force.

Forces are vector valued quantities, which means that if two forces are acting on the same body, the
resulting force, that accelerates the body, is the vector sum of the two single force vectors. And conversely
it is possible to write a given force vector as the vector sum of two other force vectors. In the case of the
inclined plane one can write the gravitational force as the vector sum of the slope force and the normal force.
The normal force $\vec{F_n}$ is the force perpendicular to the inclined plane, that holds the body on the inclined
plane (see figure 1):

<p style="text-align:center">$$ \vec{F_G}= \vec{F_sl} + \vec{F_n} $$
  <p style="text-align:right"> (6) </p>


For the length of the slope vector and the length of the normal vector follows then:

<p style="text-align:center">$$F_sl=m g sin α$$ 
and 
$$F_n=mg cos α$$
  <p style="text-align:right"> (7) </p>



The slope force is the accelerating force, because it points in the direction of the acceleration, so that the
scalar version of Newton’s second law becomes $F_sl = ma$, or with $mg sin α = ma$. This leads then to
the following formula for the acceleration down the inclined plane:

<br> <br>
<div style="border: 2px solid royalblue">
<br> <p style="text-align:center"> $$a = g sin α $$   <p style="text-align:right"> (8) <br> </p> </div>
<br>


Note, that in the limit $α = 90^\circ $ one has again the free fall case with $a = g$.
